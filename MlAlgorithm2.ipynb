```python
import numpy as np
import matplotlib.pyplot as plt

def linear_regression_1(n_sample=100, LR=0.01, epochs=1000):
    '''
    linear regression for learn more
    '''
    #np.random.seed(7)
    x = np.random.rand(n_sample) * 10
    y = 2 * x + np.random.rand(n_sample) * 10


    # Initialize weight and learning rate
    b0 = 0
    b1 = 0




    mae_ghabli = 1000
    for epoch in range(epochs):
        # Predictions for all x values
        ypred = b0 + (x * b1)

        # Errors for all x values
        err = y - ypred

        # Update weight based on the average error multiplied by the average x value.
        b1 += LR * np.mean(err * x)
        b0 += LR * np.mean(err)

        # Calculate Mean Absolute Error (MAE)
        mae_hazer = np.mean(np.abs(err))

        if epoch % 200 == 0:
            print(f'epoch: {epoch}', f'b0: {round(b0, 2)}', f'b1: {round(b1, 2)}', f'mae: {round(mae_hazer, 4)}', sep=' \t---\t')



        if abs(mae_ghabli - mae_hazer) < 0.0001:
            print(f'>>>> epoch: {epoch}', f'mae: {mae_hazer}', sep='\t\t\t')
            break

        mae_ghabli = mae_hazer




    plt.scatter(x, y, color='k', label='Real Data Points')
    plt.plot(x, ypred, color='red', label='Regression Line')
    plt.legend()
    plt.title('REGRESSION')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.grid()
    plt.show()

    return ypred, err, b0, b1
```

```python
################################################################################
########################################################## unsupervised learning
#######################################################################  k means
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Define the dataset
X = np.array(
    [
        [2.7810836, 3.550537003],
        [1.465489372, 1.362125076],
        [4.396561688, 0.400293529],
        [0.7810836, 2.550537003],
        [1.465489372, 2.362125076],
        [4.396561688, 4.400293529],
        [1.38807019, 1.850220317],
        [3.06407232, 3.005305973],
        [7.627531214, 8.759262235],
        [5.332441248, 4.088626775],
        [6.922596716, 3.77106367],
        [8.675418651, -0.242068655],
        [7.673756466, 3.508563011],
        [7.122596716, 3.97106367],
        [7.675418651, -0.242068655],
        [5.673756466, 2.508563011],
        [2.1, 3.2],
        [1.9, 2.1],
        [4.5, 0.6],
        [0.9, 2.8],
        [1.7, 2.5],
        [4.8, 4.9],
        [1.4, 1.9],
        [3.2, 3.3],
        [7.9, 8.9],
        [5.6, 4.3],
        [7.1, 3.9],
        [8.9, -0.1],
        [7.8, 3.6],
        [7.3, 4.1],
        [7.9, -0.1],
        [5.8, 2.7],
    ]
)

# Define the K-Means clustering class
class KMeans:
    def __init__(self, k):
        self.k = k

    def initialize_centroids(self, X):
        np.random.seed(0)
        indices = np.random.choice(X.shape[0], self.k, replace=False)
        centroids = X[indices, :]
        return centroids

    def closest_centroid(self, X, centroids):
        distances = np.sqrt(((X - centroids[:, np.newaxis])**2).sum(axis=2))
        closest = np.argmin(distances, axis=0)
        return closest

    def update_centroids(self, X, closest, centroids):
        return np.array([X[closest==i].mean(axis=0) for i in range(self.k)])

    def fit(self, X):
        centroids = self.initialize_centroids(X)
        for _ in range(100):
            closest = self.closest_centroid(X, centroids)
            centroids = self.update_centroids(X, closest, centroids)
        return centroids, closest

# Create an instance of the KMeans class
kmeans = KMeans(k=3)

# Fit the model to the data
centroids, labels = kmeans.fit(X)

# Plot the data with the cluster labels
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.5)
plt.show()

# Evaluate the clustering using metrics
silhouette = silhouette_score(X, labels)
calinski_harabasz = calinski_harabasz_score(X, labels)
davies_bouldin = davies_bouldin_score(X, labels)

print("Silhouette Coefficient:", silhouette)
print("Calinski-Harabasz Index:", calinski_harabasz)
print("Davies-Bouldin Index:", davies_bouldin)
```

```python
################################################################################
############################################################ supervised learning
########################################################    logistic regression
import random
import numpy as np
import matplotlib.pyplot as plt

# Define the dataset with additional points
X = np.array(
    [
        [2.7810836, 3.550537003, 0],
        [1.465489372, 1.362125076, 0],
        [4.396561688, 0.400293529, 0],
        [0.7810836, 2.550537003, 0],
        [1.465489372, 2.362125076, 0],
        [4.396561688, 4.400293529, 0],
        [1.38807019, 1.850220317, 0],
        [3.06407232, 3.005305973, 0],
        [7.627531214, 8.759262235, 1],
        [5.332441248, 4.088626775, 1],
        [6.922596716, 3.77106367, 1],
        [8.675418651, -0.242068655, 1],
        [7.673756466, 3.508563011, 1],
        [7.122596716, 3.97106367, 1],
        [7.675418651, -0.242068655, 1],
        [5.673756466, 2.508563011, 1],
        [2.1, 3.2, 0],
        [1.9, 2.1, 0],
        [4.5, 0.6, 0],
        [0.9, 2.8, 0],
        [1.7, 2.5, 0],
        [4.8, 4.9, 0],
        [1.4, 1.9, 0],
        [3.2, 3.3, 0],
        [7.9, 8.9, 1],
        [5.6, 4.3, 1],
        [7.1, 3.9, 1],
        [8.9, -0.1, 1],
        [7.8, 3.6, 1],
        [7.3, 4.1, 1],
        [7.9, -0.1, 1],
        [5.8, 2.7, 1],
    ]
)

# Split the dataset into features (x1, x2) and target variable (y)
x1, x2, y = X[:, 0], X[:, 1], X[:, 2]

# Create a column of ones for the bias term
a = np.ones(len(x2))

# Combine the features into a single array
Xd = np.array(list(zip(x1, x2)))

# Convert the target variable to integers
y = [int(x) for x in y]



# Create a figure with two subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Define a colormap for plotting
colormap = np.array(["r", "g"])

# Plot the original data with their class
ax[0].scatter(x1, x2, color=colormap[y])
ax[0].set_title('Original Data with their Class')

# Define a class for logistic regression
class LR:
    def __init__(self):
        pass

    # Define the sigmoid function
    def sigmoid(self, w, x):
        # Compute the weighted sum
        res = w[0] + (np.dot(w[1:], x))
        # Apply the sigmoid function
        out = 1 / (1 + np.exp(-res))
        return out

    # Define the updater function for gradient descent
    def updater2(self, w, alpha, y, ypred, x):
        wnew = []
        for i in range(len(w)):
            if i == 0:  # Update the bias term
                w1 = w[i] + alpha * (y - ypred) * ypred * (1 - ypred) * 1
                wnew.append(w1)
            elif i >= 1:  # Update the weights
                w1 = (
                    w[i] + alpha * (y - ypred) * x[(i - 1)]
                )
                wnew.append(w1)
        return wnew

    # Define the updater function for gradient descent (alternative implementation)
    def updater(self, w, alpha, y, ypred, x):
        w1 = w[0] + alpha * (y - ypred) * ypred * (1 - ypred) * 1
        w2 = w[1] + alpha * (y - ypred) * ypred * (1 - ypred) * x[0]
        w3 = w[2] + alpha * (y - ypred) * ypred * (1 - ypred) * x[1]
        wnew = list([w1, w2, w3])
        return wnew

    # Define the function to perform gradient descent and sigmoid function continually
    def doo(self, inw, x, y, ite, lr):
        t = 0
        w = inw
        while t < ite:
            random_index = random.randint(0, len(x) - 1)
            res = self.sigmoid(w, x[random_index])
            w = self.updater2(w, lr, y[random_index], res, x[random_index])
            t += 1
        return w

    # Define the function to predict the target variable
    def predicter(self, w, x, y):
        sto = []
        for i in x:
            a = self.sigmoid(w, i)
            sto.append(a)
        res = []
        for i in sto:
            if i >= 0.5:
                i = 1
                res.append(i)
            elif i < 0.5:
                i = 0
                res.append(i)
        acc = [(i, j) for i, j in zip(y, res)]
        score = 0
        for i, j in acc:
            if i == j:
                score += 1
        print(f'Pair of true y and predicted y: {acc}', "\nAccuracy: ", score / len(x))
        return res

    # Define the function to present the results
    def present(self, index, initweight, x, y, iteration, LR):
        inw = initweight
        www = self.doo(initweight, x, y, iteration, LR)
        self.predicter(www, x, y)
        print("Weights: ", www)


# Create an instance of the LR class
sg = LR()

# Initialize the weights
inw = [0, 0, 0]

# Perform gradient descent and predict the target variable
www = sg.doo(inw, Xd, y, 2000, 0.2)
ynew = sg.predicter(www, Xd, y)

# Print the predicted weights
print(f'Predicted weights: {www}')

# Plot the predicted data
ax[1].scatter(x1, x2, color=colormap[ynew])
ax[1].set_title('Predicted Data')

# Show the plot
plt.show()
```

```python
################################################################################
############################################################## Linear regression
####################################################################    5 Num
# Training data
X = [1, 2, 3, 4, 6]
y = [2, 3, 5, 7, 11]

# Initialize weights
b0 = 0.1  # Bias
b1 = 0.1  # Slope

# Learning rate
alpha = 0.01

# Train the model
for epoch in range(100):
    # Forward pass for all training examples
    output = [b0 + b1 * x for x in X]  # Include the bias term
    error = [yi - yi_hat for yi, yi_hat in zip(y, output)]

    # Calculate and print metrics every 10 epochs
    if epoch % 30 == 0:
        mae = sum(abs(err) for err in error) / len(error)
        rmse = (sum(err ** 2 for err in error) / len(error)) ** 0.5
        print(f"Epoch {epoch}:")
        print(f"MAE: {mae:.4f}")
        print(f"RMSE: {rmse:.4f}\n")

    # Update weights using mean error across all examples
    b0 += alpha * (sum(error) / len(error))  # Update bias
    b1 += alpha * (sum(err * x for err, x in zip(error, X)) / len(error))

# Test the model
X_test = [5]
y_test = [11]

# Generate predictions
test_output = b0 + b1 * X_test[0]  # Include the bias in prediction
train_output = [b0 + b1 * x for x in X]  # Include the bias in training output

# Calculate final metrics on training data
final_mae = sum(abs(yi - yi_hat) for yi, yi_hat in zip(y, train_output)) / len(y)
final_rmse = (sum((yi - yi_hat) ** 2 for yi, yi_hat in zip(y, train_output)) / len(y)) ** 0.5

# Calculate R-squared
ss_total = sum((yi - sum(y) / len(y)) ** 2 for yi in y)
ss_residual = sum((yi - yi_hat) ** 2 for yi, yi_hat in zip(y, train_output))
r_squared = 1 - (ss_residual / ss_total)

print("Final Results:")
print(f"Input: {X_test[0]}, Output: {test_output:.2f}")
print(f"Expected: {y_test[0]}")
print(f"Final Training MAE: {final_mae:.4f}")
print(f"Final Training RMSE: {final_rmse:.4f}")
print(f"R-squared: {r_squared:.4f}\n")

print("Training Data:")
print("X\tActual\tPredicted")
for i in range(len(X)):
    print(f"{X[i]}\t{y[i]}\t{train_output[i]:.2f}")

print("\nTest Data:")
print("X\tActual\tPredicted")
print(f"{X_test[0]}\t{y_test[0]}\t{test_output:.2f}")
```
